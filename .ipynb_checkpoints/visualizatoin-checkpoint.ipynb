```python
import pandas as pd
from bokeh.io import output_notebook, show
output_notebook()
```

```python
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.plotting import figure
from bokeh.transform import factor_cmap
games = pd.read_csv('vgsales.csv')
game_data = games.sort_values(by='Global_Sales', ascending=False)[:5]
```

```python
game_data.Name = game_data.Name.astype(str)
game_data.Global_Sales = game_data.Global_Sales.astype(str)
```

```python
group = game_data.groupby(by=['Name', 'Global_Sales'])
source = ColumnDataSource(group)
```

```python
p = figure(plot_width=800, plot_height=300, title="Mean Sales by Name of Game",
           x_range=group, toolbar_location=None, tools="")
```

```python
p.xgrid.grid_line_color = None
p.xaxis.axis_label = "Global Sales"
p.xaxis.major_label_orientation = 1.2
```

```python
index_cmap = factor_cmap('Global_Sales', palette=['#2b83ba', '#abdda4', '#ffffbf', '#fdae61', '#d7191c'], 
                         factors=sorted(game_data.Name.unique()), end=1)
```

```python
p.vbar(x='Global_Sales', top='mpg_mean', width=1, source=source,
       line_color="white", fill_color=index_cmap, 
       hover_line_color="darkgrey", hover_fill_color=index_cmap)
```

```python
show(p)
```
